#Triangulo logico

```python
#entrada

a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))

#processamento

condicao = (a <(b + c))and(b <(a + c))and(c <(a + b))
equi = (a==b==c) and condicao
iso = ((a==b and a!=c) or (b==c and b!=a) or (c==a and c!=b)) and condicao
esca = (a!=b!=c!=a) and condicao



#saida

print("É um triangulo?",condicao)
print("O triangulo é equilatero?",equi)
print("O triangulo é isoceles?",iso)
print("O triangulo é escaleno?",esca)
```
